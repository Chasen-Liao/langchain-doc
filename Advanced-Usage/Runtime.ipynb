# Runtime 运行时

## 概述

Langchain的 `create_agent` 在底层是运行在 langgraph 的 Runtime 上的

langgraph 暴露了一个 Runtime 对象，包括以下信息：
- Context：静态信息，比如说用户ID、数据库连接、或者agent调用中的依赖
- Store：一个 `BaseStore` 实例，用于长期记忆
- Stream Writer：通过 `custom` 流式传输信息的对象

运行时上下文为您的工具和中间件提供依赖注入。您无需硬编码值或使用全局状态，而是在调用您的Agent时注入 `运行时依赖项`（如数据库连接、用户 ID 或配置）。这使得您的工具更易于测试、更可重用且更灵活。

一般在工具和中间件中访问 Runtime Info
